```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_it(eps):
    n = 50
    q = 2
    c = 5
    p_in = c/n - eps/(2*n)
    p_out = c/n + eps/(2*n)
    g = nx.planted_partition_graph(q, int(n/q), p_in, p_out)
    nx.draw(g)
```

```python
plot_it(0)
```

```python
plot_it(4)
```

```python
plot_it(8)
```
