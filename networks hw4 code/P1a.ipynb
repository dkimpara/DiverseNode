```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_it(eps):
    n = 50
    q = 2
    c = 5
    p_in = c/n + eps/(2*n)
    p_out = c/n - eps/(2*n)
    g = nx.planted_partition_graph(q, int(n/q), p_in, p_out)
  
    b = [v for v in range(int(n/2))]
    r = [v+n/2 for v in range(int(n/2))]
    
    pos=nx.spring_layout(g)
    nx.draw_networkx_nodes(g,pos,
                           nodelist=b,
                           node_color='b')
    nx.draw_networkx_nodes(g,pos,
                           nodelist=r,
                           node_color='r')
    nx.draw_networkx_edges(g,pos,width=1.0,alpha=0.5)
    nx.draw_networkx_labels(g,pos)
    plt.show()
```

```python
plot_it(0)
```

```python
plot_it(4)
```

```python
plot_it(8)
```
